# Removing outliers in numerical columns using Z score

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(data[numerical_columns])
```

```python
z_scaled_df = pd.DataFrame(sc.transform(data[numerical_columns]),
                          columns=numerical_columns)
```

```python
z_scaled_df.bloodpressure.min()
```

```python
def check_z(x):
    return (x >= 3) or (x <= -3)
```

```python
outlier_indices = set()
for col in numerical_columns:
    temp = z_scaled_df[z_scaled_df[col].apply(check_z)]
    if temp.shape[0] != 0:
        outlier_indices.update(temp.index)
```

```python
len(outlier_indices)
```

```python
data = data.drop(index=outlier_indices)
data = data.reset_index(drop=True)
data
```

# Feature Selection of categorical columns with respect to target variable

```python
categorical_columns
```

```python
from scipy.stats import f_oneway
p_vals = {}
for col in categorical_columns:
    groups = [data['claim'][data[col] == category].values for category in data[col].unique()]
    p_val = f_oneway(*groups)[1]
    p_vals[col] = p_val
p_vals
```

```python
significant_features , insignificant_features = [], []
for key, value in p_vals.items():
    if value < 0.05:
        significant_features.append(key)
    else:
        insignificant_features.append(key)
```

```python
significant_features
```

```python
insignificant_features
```

# Check and drop if independent categorical columns are correlated

```python
from scipy.stats import chi2_contingency
from itertools import combinations
```

```python
cat_combs = combinations(categorical_columns,2)
chi2_vals = {}
for comb in cat_combs:
    feat_a , feat_b = comb
    cross_tab = pd.crosstab(data[feat_a], data[feat_b])
    p_val = chi2_contingency(cross_tab)[1]
    chi2_vals[comb] = p_val
```

```python
chi2_vals
```

```python
significant_comb, insignificant_comb = [] , []
for comb, p_val in chi2_vals.items():
    if p_val <= 0.05:
        significant_comb.append(comb)
    else:
        insignificant_comb.append(comb)
```

```python
significant_comb
```

```python
insignificant_comb
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error
x = data.drop(columns='claim')
y = data['claim']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=45,stratify=data['job_title'])
```

```python
x_train.shape
```

```python
x_test.shape
```

# Frequency encoding for job_title column

```python
job_title_freq = x_train['job_title'].value_counts()
job_title_freq
```

```python
x_train['job_title'] = x_train['job_title'].map(job_title_freq)
x_train
```

```python
x_test['job_title'] = x_test['job_title'].map(job_title_freq)
x_test
```

# Onehot encoding categorical columns except job_title

```python
categorical_columns
```

```python
from sklearn.preprocessing import OneHotEncoder
train_ohe_df = pd.DataFrame()
test_ohe_df = pd.DataFrame()

for col in categorical_columns:
    if col == 'job_title':
        continue
    ohe = OneHotEncoder(sparse=False)
    ohe.fit(x_train[[col]])
    train_df = pd.DataFrame(ohe.transform(x_train[[col]]),
                      columns= ohe.get_feature_names_out(),
                           index=x_train.index)
    test_df = pd.DataFrame(ohe.transform(x_test[[col]]),
                      columns= ohe.get_feature_names_out(),
                          index=x_test.index)
    train_ohe_df = pd.concat([train_ohe_df,train_df], axis=1)
    test_ohe_df = pd.concat([test_ohe_df,test_df], axis=1)
```

```python
train_ohe_df
```

```python
test_ohe_df
```

```python
x_train_df = pd.concat([x_train[numerical_columns], x_train['job_title'],train_ohe_df,y_train], axis=1)
x_train_df
```

```python
x_test_df = pd.concat([x_test[numerical_columns], x_test['job_title'],test_ohe_df,y_test], axis=1)
x_test_df
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(x_train_df)
```

```python
x_train_scaled = pd.DataFrame(ss.transform(x_train_df),
                             index = x_train_df.index,
                             columns= x_train_df.columns)
x_test_scaled = pd.DataFrame(ss.transform(x_test_df),
                             index = x_test_df.index,
                             columns= x_test_df.columns)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
models = [LinearRegression(), DecisionTreeRegressor(), RandomForestRegressor(), KNeighborsRegressor()]
model_names = ['Linear Regression', 'Decision Tree', 'Random Forest', 'KNN']
train_r2, test_r2, train_mae, test_mae = [], [], [], []
for model_name, model in zip(model_names, models):
    model.fit(x_train_scaled, y_train)
    y_pred_train = model.predict(x_train_scaled)
    y_pred_test = model.predict(x_test_scaled)
    train_r2.append(round(r2_score(y_train, y_pred_train), 2))
    test_r2.append(round(r2_score(y_test, y_pred_test), 2))
    train_mae.append(round(mean_absolute_error(y_train, y_pred_train),2))
    test_mae.append(round(mean_absolute_error(y_test, y_pred_test), 2))
result_df = pd.DataFrame()
result_df['model_name'] = model_names
result_df['train_r2'] = train_r2
result_df['test_r2'] = test_r2
result_df['train_mae'] = train_mae
result_df['test_mae'] = test_mae
result_df
```

# Calculating adjusted r2

```python
def calculate_adj_r2(x, n, k):
    return 1 - ((1 - x) * (n - 1)) / (n - k - 1)
```

```python
result_df['test_r2'].apply(lambda x: calculate_adj_r2(x, x_test.shape[0], x_train.shape[1]))
```
